```python
from transformers import pipeline
```

```python
generator = pipeline("text-generation", model="gpt2")
```

```python
prompt = "Artificial Intelligence will change the future because"
output = generator(prompt, max_length=50, num_return_sequences=1)
print(output[0]['generated_text'])
```

# Evaluation

The model generates coherent sentences based on a given prompt.  
However, since it is pre-trained on general data, sometimes the text may not stay fully on topic.  
This demonstrates both the power and limitations of Generative AI models.
